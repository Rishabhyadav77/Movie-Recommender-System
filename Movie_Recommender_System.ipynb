```python
import ast   # string convert into list.
def convert(obj):
    l=[]
    for i in ast.literal_eval(obj):
        i['name']
        l.append(i['name'])
    return l
```

```python
movie['genres']=movie['genres'].apply(convert)
movie.head()
```

```python
movie['keywords']=movie['keywords'].apply(convert)
movie.head()
```

```python
movie['cast'][0]
```

```python
def convert3(obj):
    l=[]
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            l.append(i['name'])
            counter +=1
        else:
            break
    
    return l
```

```python
movie['cast'] = movie.cast.apply(convert3)
movie.head()
```

```python
def fetch_director(obj):
    l=[]
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            l.append(i['name'])
            break
    
    return l
```

```python
movie['crew']=movie['crew'].apply(fetch_director)
movie.head()
```

```python
movie['overview']=movie['overview'].apply(lambda x:x.split()) # sring convert into list beacuse it  is easy to concatnate with other list
movie.head()
```

```python
#[Sam Mendes]
#[Sam Worthington] if you  search sam mendes but recommender show sam  Worthington so it is wrong se we can remove the space
# in this names. now the name like samWorthington so there is no error by recommender.
```

```python
movie['genres']=movie['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movie['keywords	']=movie['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movie['cast']=movie['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movie['crew']=movie['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movie.head()
```

```python
movie['tags']=movie['overview'] + movie['cast'] + movie['genres'] + movie['keywords'] +movie['crew']
```

```python
new_df= movie[['movie_id' , 'title', 'tags' ]]
```

```python
new_df
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
new_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv= CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps= PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    
    return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
#we can use cosine to check the closure of movies
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
similarity
```

```python
similarity[1]
```

```python
similarity = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title']== movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
     
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
        
```

```python
recommend('Batman Begins')
```
